```python
import numpy as np
import seaborn as sns
import pandas as pd
pdf = pd.DataFrame(sns.load_dataset('mpg'))
print(pdf.dtypes)
pdf = pdf.dropna()
pdf = pdf.drop_duplicates(subset='name', keep="first")
print(pdf)
```

```python
midpoint = len(pdf)//2 
t_df = pdf[['horsepower','weight']][0:midpoint]
```

```python
class mather:
  offset = 0
  def __init__(self,n):
    self.offset = n

  def offsetter(self,n):
    n += self.offset
    return n
```

```python
class proxy(mather):
  def divisibility(self,num):
    if(type(num) == int or type(num) == float):
      num  = self.offsetter(num)
      if(num % 2 == 0):
        return 'even'
      else:
        return 'odd'
    else:
      return np.nan

tester = proxy(1)
tester.divisibility(28)
t_df['horsepower offset evenness'] = t_df['horsepower'].apply(tester.divisibility)
t_df['weight offset evenness'] = t_df['weight'].apply(tester.divisibility)
print(t_df)
```
